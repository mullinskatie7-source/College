## Problem Description

This assignment uses Python control flow and data structures to analyze structural load readings from bridge sensors. The program will validate readings, remove invalid sensor values, classify valid loads into engineering categories, and summarize the results. This helps demonstrate how Python can be used as a practical engineering tool for processing and interpreting data.

## Variable Dictionary

| Variable Name | Data Type | Description |
|---|---|---|
| `sensor_readings` | list of tuples | Stores all raw sensor readings from the bridge monitoring system. |
| `sensor_id` | int | Unique number identifying each sensor reading. |
| `node` | str | Bridge node location where the reading was collected. |
| `load` | float | Load measurement in kilonewtons (kN). |
| `valid_readings` | list of tuples | Stores only readings with non-negative load values. |
| `invalid_count` | int | Counts the number of invalid readings with negative loads. |
| `classification_report` | dictionary | Stores each valid sensor reading with its node, load, and classification status. |
| `safe_loads` | list | Stores load values classified as Safe. |
| `caution_loads` | list | Stores load values classified as Caution. |
| `critical_loads` | list | Stores load values classified as Critical. |
| `critical_nodes` | set | Stores unique node labels that had at least one Critical reading. |


```python
sensor_readings = [ (1, "Node-A", 102.3), (2, "Node-B", 178.5), (3, "Node-C", 263.0), (4, "Node-D", -15.2), (5, "Node-E", 149.9), (6, "Node-A", 251.7), (7, "Node-B", 88.4), (8, "Node-C", -3.0), (9, "Node-D", 199.6), (10, "Node-E", 305.1), (11, "Node-A", 145.0), (12, "Node-B", 250.0), (13, "Node-C", 73.2), (14, "Node-D", 310.8), (15, "Node-E", 160.4), ]

# Verify: print total record count and the full list

print(f"Total records loaded: {len(sensor_readings)}")
for reading in sensor_readings:
    print(reading)
```

## Step 2 — Validate and Filter Readings

In this step, I use a `for` loop to go through each sensor reading in the dataset. Each reading is a tuple, so I unpack it into `sensor_id`, `node`, and `load`. If the load value is negative, the reading is invalid because a negative structural load represents a sensor error. When an invalid value is found, the program prints a warning, increases the invalid count, and uses `continue` to skip the rest of that loop cycle. Valid readings are stored in a new list called `valid_readings` so they can be used in the later classification steps.

```python
# Step 2: Validate and filter sensor readings.
# Invalid readings are negative load values, so they are skipped.

valid_readings = []
invalid_count = 0

for sensor_id, node, load in sensor_readings:
    if load < 0:
        print(f"WARNING: Sensor {sensor_id} ({node}) recorded an invalid reading of {load} kN. Skipping.")
        invalid_count += 1
        continue

    valid_readings.append((sensor_id, node, load))

print(f"\nValid readings: {len(valid_readings)}")
print(f"Invalid readings: {invalid_count}")
```

## Step 3 — Classify Each Valid Reading

In this step, I classify each valid sensor reading based on the given engineering load ranges. I use an `if`, `elif`, and `else` structure because each valid reading can only belong to one category. Loads less than 150 kN are classified as Safe, loads from 150 kN to less than 250 kN are classified as Caution, and loads greater than or equal to 250 kN are classified as Critical. I store the classified results in a dictionary using the sensor ID as the key, and I also store the numeric load values in separate category lists for later summary calculations.

```python
# Step 3: Classify each valid reading and store results.
# A dictionary stores the full report, while separate lists store values for statistics.

classification_report = {}
safe_loads = []
caution_loads = []
critical_loads = []

for index, (sensor_id, node, load) in enumerate(valid_readings):
    if load < 150:
        status = "Safe"
        safe_loads.append(load)
    elif load < 250:
        status = "Caution"
        caution_loads.append(load)
    else:
        status = "Critical"
        critical_loads.append(load)

    classification_report[sensor_id] = {
        "node": node,
        "load_kN": load,
        "status": status
    }

print("=== Classification Report ===")

for sid, record in classification_report.items():
    print(f"Sensor {sid:>2} | {record['node']} | {record['load_kN']:>6.1f} kN | {record['status']}")
```

## Step 4a — Category Statistics

In this step, I summarize the Safe, Caution, and Critical load categories. For each category, I calculate the number of readings, minimum load, maximum load, and average load. Before using `min()`, `max()`, or an average calculation, I check whether the list contains values. This prevents an error if one of the categories is empty.

```python
# Step 4a: Compute and print statistics for each classification category.
# Empty-list checks prevent errors if a category has no readings.

categories = {
    "Safe": safe_loads,
    "Caution": caution_loads,
    "Critical": critical_loads
}

print("=== Category Statistics ===")

for category, loads in categories.items():
    print(f"\n{category} Readings")
    print(f"Count: {len(loads)}")

    if len(loads) > 0:
        minimum = min(loads)
        maximum = max(loads)
        average = sum(loads) / len(loads)

        print(f"Minimum Load: {minimum:.1f} kN")
        print(f"Maximum Load: {maximum:.1f} kN")
        print(f"Average Load: {average:.1f} kN")
    else:
        print("No readings available for this category.")
```

## Step 4b — Critical Node Identification

In this step, I use a set to collect the unique node labels that had at least one Critical reading. A set is useful because it automatically removes duplicates. This matters because the same node can appear more than once in the dataset, but the summary should only list each critical node one time.

```python
# Step 4b: Collect unique node labels that have at least one Critical reading.
# A set is used because it automatically removes duplicate node names.

critical_nodes = set()

for sensor_id, record in classification_report.items():
    if record["status"] == "Critical":
        critical_nodes.add(record["node"])

print("=== Critical Nodes ===")

if len(critical_nodes) > 0:
    for node in sorted(critical_nodes):
        print(node)
else:
    print("No critical nodes found.")
```

## Step 4c — Threshold Scan with Early Exit

In this step, I scan the original sensor readings to find the first load value that is greater than or equal to 300 kN. I use a `for` loop with `break` because once the first matching reading is found, the program does not need to keep searching. I also use the loop `else` statement, which only runs if the loop finishes without finding a matching value.

```python
# Step 4c: Scan for the first reading greater than or equal to 300 kN.
# The break statement stops the loop as soon as the first match is found.

print("=== First Reading At or Above 300 kN ===")

for sensor_id, node, load in sensor_readings:
    if load >= 300:
        print(f"First reading >= 300 kN: Sensor {sensor_id} ({node}) with {load:.1f} kN")
        break
else:
    print("No reading was greater than or equal to 300 kN.")
```
