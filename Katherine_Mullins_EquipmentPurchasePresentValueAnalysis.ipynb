## Problem Description

The purpose of this analysis is to determine the best equipment purchase scenario using present value calculations. The equipment has an original purchase price of $10,000 and a 10-year life span. Each purchase scenario has a different resale value at year 10 and a different guaranteed annual revenue.

The analysis will calculate the present value of the future resale amount, the present value of the annual revenue stream, the total present value, and the profit or loss compared to the original $10,000 purchase price. The base analysis uses a 5% interest rate, and additional calculations are performed at lower and higher interest rates to evaluate risk.

## Methodology

The present value of the resale amount is calculated by discounting the future resale value back 10 years using the interest rate.

PV of Resale = Future Resale Value / (1 + interest rate)^n

The present value of the revenue stream is calculated by discounting each annual revenue payment back to present value and adding the yearly values together.

PV of Revenue = Sum of [Annual Revenue / (1 + interest rate)^year] for years 1 through 10

The total present value is:

Total PV = PV of Resale + PV of Revenue

The profit or loss is:

Profit or Loss = Total PV - Original Purchase Price


```python
# Step 2: Import math and define constants
import math

# Given values
purchase_price = 10000
n = 10  # equipment life in years

# Interest rates for risk analysis
# 0.05 is the required base case.
interest_rates = [0.03, 0.05, 0.07]

# Step 3: Equipment scenarios
resale_values = [10000, 9000, 5000, 3000, 0]
annual_revenues = [500, 600, 1000, 1200, 1500]

# Step 4: Define present value functions
# Function to calculate present value of future resale value
def calculate_pv_resale(resale_amount, n, interest):
    pv_resale = resale_amount / math.pow(1 + interest, n)
    return pv_resale


# Function to calculate present value of annual revenue
# This uses the professor's method: discount each year one at a time and add them.
def calculate_pv_revenue(annual_revenue, n, interest):
    pv_revenue = 0

    for year in range(1, n + 1):
        yearly_value = annual_revenue / math.pow(1 + interest, year)
        pv_revenue += yearly_value

    return pv_revenue


# Step 5: Run the analysis for each interest rate
for interest in interest_rates:
    print()
    print("=" * 90)
    print(f"Present Value Analysis at {interest * 100:.1f}% Interest")
    print("=" * 90)

    print(f"{'Resale Amount':>15} | {'PV of Resale':>15} | {'Annual Revenue':>15} | "
          f"{'PV of Revenue':>15} | {'Total PV':>12} | {'Profit/Loss':>15}")
    print("-" * 105)

    index = 0
    best_total_pv = -1
    best_scenario = None

    # While loop processes each equipment scenario
    while index < len(resale_values):
        resale_amount = resale_values[index]
        annual_revenue = annual_revenues[index]

        pv_resale = calculate_pv_resale(resale_amount, n, interest)
        pv_revenue = calculate_pv_revenue(annual_revenue, n, interest)
        total_pv = pv_resale + pv_revenue
        profit_loss = total_pv - purchase_price

        print(f"${resale_amount:>14,.2f} | ${pv_resale:>14,.2f} | "
              f"${annual_revenue:>14,.2f} | ${pv_revenue:>14,.2f} | "
              f"${total_pv:>11,.2f} | ${profit_loss:>14,.2f}")

        # Track the best scenario for this interest rate
        if total_pv > best_total_pv:
            best_total_pv = total_pv
            best_scenario = (resale_amount, annual_revenue, total_pv, profit_loss)

        index += 1

    print("-" * 105)
    print(f"Best scenario at {interest * 100:.1f}% interest:")
    print(f"Resale Amount: ${best_scenario[0]:,.2f}")
    print(f"Annual Revenue: ${best_scenario[1]:,.2f}")
    print(f"Total PV: ${best_scenario[2]:,.2f}")
    print(f"Profit/Loss: ${best_scenario[3]:,.2f}")
```
